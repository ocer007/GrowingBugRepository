```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from random import randint


# txt -> pandas
facebook = pd.read_csv(
    "/content/sample_data/facebook_combined.txt.gz",
    compression="gzip",
    sep=" ",
    names=["start_node", "end_node"],
)

# pandas->network
G = nx.from_pandas_edgelist(facebook, "start_node", "end_node")

# 
def show_graph():
    pos = nx.spring_layout(G, iterations=15, seed=1721)
    fig, ax = plt.subplots(figsize=(15, 9))
    ax.axis("off")
    plot_options = {"node_size": 10, "with_labels": False, "width": 0.15}
    nx.draw_networkx(G, pos=pos, ax=ax, **plot_options)
    plt.show()
    return
# 1 Average path length
#  Average path length
def average_path_length():
    result = nx.average_shortest_path_length(G)
    return result
# 
def show_path_length_table():
    # 
    shortest_path_lengths = dict(nx.all_pairs_shortest_path_length(G))
    # 
    diameter = max(nx.eccentricity(G, sp=shortest_path_lengths).values())

    path_lengths = np.zeros(diameter + 1, dtype=int)
    for pls in shortest_path_lengths.values():
        pl, cnts = np.unique(list(pls.values()), return_counts=True)
        path_lengths[pl] += cnts
    freq_percent = 100 * path_lengths[1:] / path_lengths[1:].sum()
    print(type(freq_percent))
    fig, ax = plt.subplots(figsize=(15, 8))
    ax.bar(np.arange(1, diameter + 1), height=freq_percent)
    ax.set_title(
        "Distribution of shortest path length in G", fontdict={"size": 35}, loc="center"
    )
    ax.set_xlabel("Shortest Path Length", fontdict={"size": 22})
    ax.set_ylabel("Frequency (%)", fontdict={"size": 22})
    # plt.show()
    return

# 2 Degree distribution
# 
def average_degree():
    # 
    dict_degree = dict(nx.degree(G))
    result = sum(dict_degree.values()) / len(dict_degree)
    return result
# 
def show_degree_distribution():
    # 
    hist_degree = nx.degree_histogram(G)
    dict_degree = dict(nx.degree(G))
    x = list(range(max(dict_degree.values()) + 1))
    y = [i*100/sum(hist_degree) for i in hist_degree]
    plt.figure(figsize=(15, 8))
    plt.bar(x,y)
    plt.xlabel("Degree", fontdict={"size": 22})
    plt.ylabel("Frequency (%)", fontdict={"size": 22})
    plt.xlim([0,max(dict_degree.values())+1])
    plt.show()
    return

# 3 clustering coefficient
# 
def average_clustering_G():
    result = nx.average_clustering(G)
    return result

# 
def show_clustering_distribution():
    plt.figure(figsize=(15, 8))
    plt.hist(nx.clustering(G).values(), bins=25)
    plt.title("Clustering Coefficient Histogram ", fontdict={"size": 35}, loc="center")
    plt.xlabel("Clustering Coefficient", fontdict={"size": 20})
    plt.ylabel("Counts", fontdict={"size": 20})
    plt.show()
    return


# 4 Assortativity
# 
def assortativity_G():
    result = nx.degree_assortativity_coefficient(G)
    return result

# 5 Network community
# LPA
def LPA_for_network_community():
    label_propagation_communities = nx.community.label_propagation_communities(G)
    counter = 0
    colors = ["" for x in range(G.number_of_nodes())]
    for com in label_propagation_communities:
        color = '#%06X' % randint(0, 0xFFFFFF)  # creates random RGB color
        counter += 1
        for node in list(com):  # fill colors list with the particular color for the community nodes
            colors[node] = color

    pos = nx.spring_layout(G, iterations=15, seed=1721)
    plt.figure(figsize=(15, 9))
    title = "%d communities found in total with LPA" % counter
    plt.title(title, fontdict={"size": 20}, loc="center")
    plt.axis('off')
    nx.draw_networkx(G, pos=pos, node_size=10, with_labels=False, width=0.15, node_color=colors)
    plt.show()
    return
# greedy_modularity
def greedy_modularity_for_network_community():
    greedy_modularity_communities = nx.community.greedy_modularity_communities(G)
    counter = 0
    colors = ["" for x in range(G.number_of_nodes())]
    for com in greedy_modularity_communities:
        color = '#%06X' % randint(0, 0xFFFFFF)  # creates random RGB color
        counter += 1
        for node in list(com):  # fill colors list with the particular color for the community nodes
            colors[node] = color

    pos = nx.spring_layout(G, iterations=15, seed=1721)
    plt.figure(figsize=(15, 9))
    title = "%d communities found in total with greedy_modularity" % counter
    plt.title(title, fontdict={"size": 20}, loc="center")
    plt.axis('off')
    nx.draw_networkx(G, pos=pos, node_size=10, with_labels=False, width=0.15, node_color=colors)
    plt.show()
    return
# Louvain Fast community unfolding
def fast_community_unfoldingy_for_network_community():
    fast_community_unfoldingy_communities = nx.community.louvain_communities(G)
    counter = 0
    colors = ["" for x in range(G.number_of_nodes())]
    for com in fast_community_unfoldingy_communities:
        color = '#%06X' % randint(0, 0xFFFFFF)  # creates random RGB color
        counter += 1
        for node in list(com):  # fill colors list with the particular color for the community nodes
            colors[node] = color

    pos = nx.spring_layout(G, iterations=15, seed=1721)
    plt.figure(figsize=(15, 9))
    title = "%d communities found in total with Louvain Fast community unfolding" % counter
    plt.title(title, fontdict={"size": 20}, loc="center")
    plt.axis('off')
    nx.draw_networkx(G, pos=pos, node_size=10, with_labels=False, width=0.15, node_color=colors)
    plt.show()
    return
# asyn_fluidc
def asyn_fluidc_for_network_community(counter):
    asyn_fluidc_communities = nx.community.asyn_fluidc(G,counter);
    colors = ["" for x in range(G.number_of_nodes())]
    for com in asyn_fluidc_communities:
        color = "#%06X" % randint(0, 0xFFFFFF)  # creates random RGB color
        for node in list(com):
            colors[node] = color
    pos = nx.spring_layout(G, iterations=15, seed=1721)
    plt.figure(figsize=(15, 9))
    title = "%d communities found in total with Louvain Fast community unfolding" % counter
    plt.title(title, fontdict={"size": 20}, loc="center")
    plt.axis('off')
    nx.draw_networkx(G, pos=pos, node_size=10, with_labels=False, width=0.15, node_color=colors)
    plt.show()
    return









# 1  Average path length
# 
print("average_path_length:",average_path_length())
# 
show_path_length_table()

# 2 Degree distribution
# 
print("average_degree:",average_degree())
# 
show_degree_distribution()

# 3 Clustering Coefficient
# 
print("average_clustering:",average_clustering_G())
# 
show_clustering_distribution()

# 4 Assortativity
# 
print("assortativity: ",assortativity_G())

# 5 Network community
# LPA 
LPA_for_network_community()
# greedy_modularity
greedy_modularity_for_network_community()
# Louvain Fast community unfolding
# fast_community_unfoldingy_for_network_community()
# asyn_fluidc
asyn_fluidc_for_network_community(10)


# 6 Visualize the whole graph connectivity.
# 
print(nx.number_connected_components(G))
# 
show_graph()


```
